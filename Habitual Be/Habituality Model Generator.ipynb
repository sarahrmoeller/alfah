```python
# Import dataset

preData = pd.read_csv("test rules coraal_analysis_spreadsheet.csv")
```

```python
# Creates a Sentence object
# habituality is set to 0, but if the tagger finds instance of person number disagreement, will later be set to 1
# pos1...synPar4 are boolean variables for the rules, e.g '0' if conditions are not met, '1' if conditions are met
# text = text of Sentence

class Sentence:
    habituality = 0 #using a ternary system: -1 = nonhabitual 0 = unclassified, 1 = habitual
    pos1 = 0; pos2 = 0; pos3 = 0; pos4 = 0; pos5 = 0; pos6 = 0; a1 = 0; a2 = 0; 
    synPar1 = 0; synPar2 = 0; synPar3 = 0; synPar4 = 0;
    
    def __init__(self, text):
        self.text = text
```

```python
# Text preprocessing

lines = []
with open("test gold standard lines+labels.csv", "r", encoding="utf_8") as file:
    lines = file.readlines()

i = 0
del lines[0]
while (i < len(lines)):
    if (lines[i] == "\"\n" or lines[i] == "\"" or lines[i] == ' "\n'):
        del lines[i]
    i +=1
```

```python
# Text preprocessing

sentences = []

for line in lines:
    if(line != '",,,,,\n'):
        sentences.append(Sentence(line))
```

```python
# Identifies 'be' in each sentence and replicates any sentence with more than one 'be' such that each 'be' can be 
# independently analyzed for habituality

sen = []
beIndicies = []
numBe = 0
extraBe = []

data = copy.deepcopy(preData)
i = len(preData)-1
while (i >= 0):
    data.drop(i, inplace=True)
    i = i -1
    
    
for Sentence in range(len(sentences)):
    numBe = 0
    parsed = nlp(sentences[Sentence].text)
    for i, word in enumerate(parsed):
        if (word.text.lower() == "be"):
            numBe += 1
    while (numBe >= 1):
        sen.append(sentences[Sentence])  
        numBe -= 1
```

```python
# Add sentences to 'data' data frame

for Sentence in range(len(sen)):
    numBe = 0
    parsed = nlp(sen[Sentence].text)
    for i, word in enumerate(parsed):
        if (word.text.lower() == "be"):
            numBe += 1
    while (numBe >= 1):
        data = pd.concat([data, preData.iloc[Sentence]], ignore_index=True)
        numBe -= 1
```

```python
# Create arrays for parts of speech, e.g. w = word = 'be', 'p' = previous word = word before 'be',
#'pp' = pre-previous word = two words before 'be', etc

w = []
pppp = []
ppp = []
pp = []
p = []
a = []
pa = []
ppa = []
pppa = []
```

```python
# Count number of 'be' in sentence

beNum = []
for Sentence in sen:
    numBe = 0
    if nlp(Sentence.text) != parsed:
        parsed = nlp(Sentence.text)
        for i, word in enumerate(parsed):
            if (word.text.lower() == "be"):
                numBe += 1
    beNum.append([Sentence.text,numBe])
```

```python
# Append parts of speech for each sentence to arrays

numBe = 1
extraBe = []

for Sentence in sen:
    if (numBe == 1):
        numBe = 0
        parsed = nlp(Sentence.text)
        for i, word in enumerate(parsed):
            if (word.text.lower() == "be"):
                numBe += 1
                w.append(word.pos_)
                beIndicies.append(i)
                beChildren = list(word.children)
                if(word.head == word):
                    beSibling = []
                else:
                    beSibling = list(word.head.children)
                beDep = word.dep_
                bePOS = word.pos_
                beIndex = i
                prepreprevious = None
                preprevious = None
                previous = None
                after = None
                postafter = None
                postpostafter = None
                if(i >= 4):
                    preprepreprevious = parsed[i-4]
                    pppp.append(preprepreprevious.pos_)
                else:
                    pppp.append(-1)    
                if(i >= 3):
                    prepreprevious = parsed[i-3]
                    ppp.append(prepreprevious.pos_)
                else:
                    ppp.append(-1)
                if(i >= 2):
                    preprevious = parsed[i-2]
                    pp.append(preprevious.pos_)
                else:
                    pp.append(-1)
                if(i > 0):
                    previous = parsed[i-1]
                    p.append(previous.pos_)
                else:
                    p.append(-1)
                if(i+1 < len(parsed)):
                    after = parsed[i+1]
                    a.append(after.pos_)
                else:
                    a.append(-1)
                if(i+2 < len(parsed)):
                    postafter = parsed[i+2]
                    pa.append(postafter.pos_)
                else:
                    pa.append(-1)
                if(i+3 < len(parsed)):
                    postpostafter = parsed[i+3]
                    ppa.append(postpostafter.pos_)
                else:
                    ppa.append(-1)
                if(i+4 < len(parsed)):
                    postpostpostafter = parsed[i+4]
                    pppa.append(postpostpostafter.pos_)
                else:
                    pppa.append(-1)
    else:
        numBe -= 1
```

```python
## Extract n-grams from each sentence

#df = pd.read_csv('spam.csv',encoding = 'ISO-8859-1',usecols=['v1','v2'])
corpus = [] #empty list
#wordnet = wordnet() #object instantiation
wlemmatizer = WordNetLemmatizer()
length = len(data['Sentence']) #finding total number of rows
be_sen = []
for i in data['Sentence']:
    start = 0
    end = 0
    if nlp(i) != parsed:
        parsed = nlp(i)
        for j, word in enumerate(parsed):
            if word.text.lower() == "be":
                if(j >= 4):
                    start = j-4
                else:
                    start = 0
                if(len(i) > j+4):
                    end = j+4 
                else:
                    end = len(i)
        be_sen.append(parsed[start:end])

for i in range(length):
	rev = re.sub('[^a-zA-Z]',' ',str(be_sen[i]))
	rev = rev.lower() #text to lowercase
	rev = rev.split() #each word of the sentence becomes the element of a list
	rev = [wlemmatizer.lemmatize(word) for word in rev if word not in stopwords.words('english')] #lemmatization via list comprehension
	rev = ' '.join(rev) #from list to string
	corpus.append(rev) #appending to the list
    
#ngram_range=(3, 3)
cv = CountVectorizer(ngram_range=(1, 2), max_features=int(length/10)) 
x = cv.fit_transform(corpus).toarray() #converting to array
y = data['Habituality'] #dependent variable
#for i in range(len(y)):
#    if ',' in y[i]:
#        y[i] = y[i][1:]

## y is a categorical variable so will encode it
#le = LabelEncoder()
#y = le.fit_transform(y)

## now splitting the model into train and test set

#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
## Train the n-gram model and print its confusion matrix

from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB() # using naive bayes classification algorithm 
predict_fit = model.fit(x,y)# fitting the model

kfold = StratifiedKFold(n_splits=10)
predictions = cross_val_predict(model, x, y, cv=kfold)
target_names = ['nonhabitual', 'habitual']

## predicting the values
#y_pred = model.predict(x_test)

#score of the model
#model.score(x_test,y_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, predictions)

print(cm)
```
