```python
import pathlib
import csv

FOLDER = 'KFolds/AugmentedOnly/'
NEWFOLDER = 'transform/augOnly/'
DEVLEN = 402
AUGDEVLEN = 774
```

# From KFold Folders

```python
# Examine pathnames and identified needed files

directory_list = [f for f in pathlib.Path(FOLDER).glob('**/*')]

# get list of file pathnames
file_list = [f for f in directory_list if f.is_file()]
print(file_list)
```

```python
#reformat data to input and output files needed for Fairseq 

for filepath_object in file_list:
    rows = []
    filepath = str(filepath_object)
    modelname = 'augOnly' + filepath.split()[2] + 'Kfold' + filepath.split()[3]
    if filepath.endswith('Expected Output.csv'):
        xput = '.output'
    else:
        xput = '.input'
    if 'Train' in filepath:
        split = 'train.'
    else:
        split = 'test.'
    
    # read data
    with open(filepath) as csvfile:
        csvobject = csv.reader(csvfile)
        for r,row in enumerate(csvobject): 
            if r > 0:
                cells = []
                for col in row[1:]:
                    cells.append(col)
                rows.append(' '.join(cells))
    
    if split == 'train.': #split dev data & write
        devdata = rows[:DEVLEN]
        traindata = rows[DEVLEN:]
        
        devfilepath = NEWFOLDER + 'dev.'+ modelname + xput
        with open(devfilepath, 'w') as new:
            new.write('\n'.join(devdata))
        
        trainfilepath = NEWFOLDER + split + modelname + xput
        with open(trainfilepath, 'w') as new:
            new.write('\n'.join(traindata))
    
    else:  # write test data
        testfilepath = NEWFOLDER + split + modelname + xput
        with open(testfilepath, 'w') as new:
            new.write('\n'.join(rows))
```

# From Augment folders

```python
# Examine pathnames and identified needed files

directory_list = [f for f in pathlib.Path(FOLDER).glob('**/*') if 'Augmented' in str(f)]
file_list = [f for f in directory_list if f.is_file()]

topfolders = ['Baseline 1', 'Baseline 2', 'Model 3', 'Model 4', 'Model 5', 'Model 6']
innerfolders = ['with baseline1', 'with baseline2', 'without baselines']
models = ['Model1', 'Model2', 'Model3', 'Model4', 'Model5', 'Model6']
inner_models = ['Baseline1', 'Baseline2', 'NoBaseline']
```

```python
for filepath_object in file_list:
    # get modelname
    modelname = 'aug'
    filepath = ''.join(str(filepath_object).split('.')[0])
    for t,topmodel in enumerate(topfolders):
        if topmodel in filepath:
            modelname += models[t]
            continue
    for i,innerfolder in enumerate(innerfolders):
        if innerfolder in filepath:
            modelname += inner_models[i]
            continue
    modelname += 'Kfold' + filepath.split('_')[-1]
    
    # get rest of filenames
    if 'Y_' in filepath:
        xput = '.output'
    else:
        xput = '.input'
    if 'train' in filepath:
        split = 'train.'
    elif 'test' in filepath:
        split = 'test.'
    
    # read data
    rows = []
    with open(str(filepath_object)) as csvfile:
        csvobject = csv.reader(csvfile)
        for r,row in enumerate(csvobject): 
            cells = []
            if r > 0:
                for col in row:
                    cells.append(col)
                rows.append(' '.join(cells))
   
    if split == 'train.':  #split dev data 
        devdata = rows[:AUGDEVLEN]
        traindata = rows[AUGDEVLEN:]
        
        devfilepath = NEWFOLDER + 'dev.' + modelname + xput
        with open(devfilepath, 'w') as new:
            new.write('\n'.join(devdata))
        
        trainfilepath = NEWFOLDER + split + modelname + xput
        with open(trainfilepath, 'w') as new:
            new.write('\n'.join(traindata))
    
    else:  # write test data
        testfilepath = NEWFOLDER + split + modelname + xput
        with open(testfilepath, 'w') as new:
            new.write('\n'.join(rows))
```
